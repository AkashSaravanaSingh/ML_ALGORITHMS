```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
data = np.array([
    [500, 10],
    [510, 11],
    [520, 12],
    [530, 13],
    [15000, 100],
    [20000, 150],
    [1500, 17],
    [1700, 34],
    [7000, 67],
    [6000, 40],
    [8000, 63],
    [13000, 78],
    [14212, 82]
])
```

```python
db = DBSCAN(eps = 1000, min_samples=2)
labels = db.fit_predict(data)
df = pd.DataFrame(data, columns=['Amount', 'Time'])
df['Label'] = labels
print("DBSAN Output: \n")
print(df)
```

```python
plt.figure(figsize=(8, 5))
colors = ['green' if label != -1 else 'red' for label in labels]
plt.scatter(data[:, 0], data[:, 1], c = colors, s = 100, edgecolors='black')
plt.xlabel('Amount')
plt.ylabel('Time')
plt.title('DBSCAN Amount Detection')
plt.grid(True)
plt.show()
```
